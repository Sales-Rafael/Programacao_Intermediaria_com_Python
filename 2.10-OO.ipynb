#### Q2.

Crie a classe `DataHora` que dará suporte ao registro de eventos de calendário.

* A classe **não possui construtor customizado**. A alteração de seu atributo se dará a partir da propriedade a seguir.
* Crie a `property` `data_hora` para manipular o atributo `_data_hora`.
    * O getter da propriedade deve retornar a data como uma string formatada (`%d/%m/%Y, %H:%M`). Use o atributo `FORMAT`. Consulte o [funcionamento do método `strftime`](https://www.programiz.com/python-programming/datetime/strftime).
    * O setter da propriedade deve receber uma string de data formatada (`%d/%m/%Y, %H:%M`) e implementar um bloco try-except que tenta converter a string em `datetime` e lança um `ValueError` caso a entrada seja inválida. * Crie o método de instância `isPassado(self)` que avalia se a `_data_hora` é menor que `datetime.now()` (a data e hora atual) e retorna `True` em caso positivo, e `False` caso contrário.
* Crie o método de instância `somaDias(self, num_dias)` que recebe um inteiro `num_dias`, soma esse valor ao atributo interno `_data_hora` e retorna a string formatada do resultado da soma (código dado a seguir).

Teste a classe `DataHora` com o seguinte código (altere o que for necessário):
```python

# definindo a data_hora através da propriedade

## editando a data_hora através da função somaDias
```

````python
#### Classe DataHora
import datetime

class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'
    
    def __init__(self):
        self._data_hora = None  
    
    @property
    def data_hora(self):
        if self._data_hora is None:
            return None
        return self._data_hora.strftime(self.FORMAT)
    
    @data_hora.setter
    def data_hora(self, value):
        try:
            self._data_hora = datetime.datetime.strptime(value, self.FORMAT)
        except ValueError:
            raise ValueError("Data/hora inválida. Use o formato DD/MM/AAAA, HH:MM")
    
    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.datetime.now()
    
    def somaDias(self, num_dias):
        if self._data_hora is None:
            raise ValueError("Nenhuma data/hora definida")
        data_hora_somada = self._data_hora + datetime.timedelta(days=num_dias)
        return data_hora_somada.str
````

````python
#### Classe EventoUnico
from abc import ABC, abstractmethod
import datetime

class EventoABC(ABC):
    def __init__(self, titulo: str, descricao: str):
        self._titulo = titulo
        self._descricao = descricao
    
    @abstractmethod
    def __str__(self) -> str:
        pass
    
    @abstractmethod
    def isConcluido(self) -> bool:
        pass

class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'
    
    def __init__(self):
        self._data_hora = None  
    
    @property
    def data_hora(self):
        if self._data_hora is None:
            return None
        return self._data_hora.strftime(self.FORMAT)
    
    @data_hora.setter
    def data_hora(self, value):
        try:
            self._data_hora = datetime.datetime.strptime(value, self.FORMAT)
        except ValueError:
            raise ValueError("Data/hora inválida. Use o formato DD/MM/AAAA, HH:MM")
    
    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.datetime.now()
    
    def somaDias(self, num_dias):
        if self._data_hora is None:
            raise ValueError("Nenhuma data/hora definida")
        data_hora_somada = self._data_hora + datetime.timedelta(days=num_dias)
        return data_hora_somada.str
    
class EventoRecorrente(EventoABC):
    def __init__(self, titulo: str, descricao: str, 
        data_hora_inicial: str, data_hora_final: str, 
        intervalo_repeticao: int):
        super().__init__(titulo, descricao)
        self._datas_hora = []
        
        data_atual = DataHora()
        data_atual.data_hora = data_hora_inicial
        
        data_final_obj = DataHora()
        data_final_obj.data_hora = data_hora_final
        
        while datetime.datetime.strptime(data_atual.data_hora, DataHora.FORMAT) <= datetime.datetime.strptime(data_final_obj.data_hora, DataHora.FORMAT):
            nova_data = DataHora()
            nova_data.data_hora = data_atual.data_hora
            self._datas_hora.append(nova_data)
            
            data_atual.data_hora = data_atual.somaDias(intervalo_repeticao)
    
    def isConcluido(self, indice: int = 0) -> bool:
        if indice < 0 or indice >= len(self._datas_hora):
            raise IndexError("Índice inválido")
        return self._datas_hora[indice].isPassado()
    
    def __str__(self) -> str:
        return "\n".join(
            f"Evento: {self._titulo}, Data: {dh.data_hora}, "
            f"Descrição: {self._descricao}, Concluído: {dh.isPassado()}"
            for dh in self._datas_hora
        )
    
    def editar_data_hora(self, data_hora_antiga: str, data_hora_nova: str):
        for dh in self._datas_hora:
            if dh.data_hora == data_hora_antiga:
                dh.data_hora = data_hora_nova
                return
        raise ValueError(f"Data/hora {data_hora_antiga} não encontrada")
````

Por fim, vamos só ver o polimorfismo em ação. Crie e preencha uma lista de eventos, sendo alguns do tipo `EventoUnico` e outros do tipo `EventoRecorrente`. Sobre essa lista, execute o laço de impressão a seguir:

A função `print` irá invocar o método especial `__str__` das classes correspondentes dependendo do tipo do objeto recebido. Aí está o polimorfismo :)

````python
from abc import ABC, abstractmethod
import datetime

class EventoABC(ABC):
    def __init__(self, titulo: str, descricao: str):
        self._titulo = titulo
        self._descricao = descricao
    
    @abstractmethod
    def __str__(self) -> str:
        pass
    
    @abstractmethod
    def isConcluido(self) -> bool:
        pass

class DataHora:
    FORMAT = '%d/%m/%Y, %H:%M'
    
    def __init__(self):
        self._data_hora = None
    
    @property
    def data_hora(self):
        if self._data_hora is None:
            return None
        return self._data_hora.strftime(self.FORMAT)
    
    @data_hora.setter
    def data_hora(self, value):
        self._data_hora = datetime.datetime.strptime(value, self.FORMAT)
    
    def isPassado(self):
        if self._data_hora is None:
            return False
        return self._data_hora < datetime.datetime.now()
    
    def somaDias(self, num_dias):
        if self._data_hora is None:
            raise ValueError("Nenhuma data/hora definida")
        data_hora_somada = self._data_hora + datetime.timedelta(days=num_dias)
        return data_hora_somada.strftime(self.FORMAT)

class EventoUnico(EventoABC):
    def __init__(self, titulo: str, descricao: str, data_str: str):
        super().__init__(titulo, descricao)
        self._data_hora = DataHora()
        self._data_hora.data_hora = data_str
    
    def isConcluido(self) -> bool:
        return self._data_hora.isPassado()
    
    def __str__(self) -> str:
        return f"Evento: {self._titulo}, Data: {self._data_hora.data_hora}, Descrição: {self._descricao}, Concluído: {self.isConcluido()}"
    
    def editar_data_hora(self, nova_data_str: str):
        self._data_hora.data_hora = nova_data_str

class EventoRecorrente(EventoABC):
    def __init__(self, titulo: str, descricao: str, data_hora_inicial: str, data_hora_final: str, intervalo_repeticao: int):
        super().__init__(titulo, descricao)
        self._datas_hora = []
        
        data_atual = DataHora()
        data_atual.data_hora = data_hora_inicial
        
        data_final = DataHora()
        data_final.data_hora = data_hora_final
        
        while datetime.datetime.strptime(data_atual.data_hora, DataHora.FORMAT) <= datetime.datetime.strptime(data_final.data_hora, DataHora.FORMAT):
            nova_data = DataHora()
            nova_data.data_hora = data_atual.data_hora
            self._datas_hora.append(nova_data)
            data_atual.data_hora = data_atual.somaDias(intervalo_repeticao)
    
    def isConcluido(self, indice: int = 0) -> bool:
        if indice < 0 or indice >= len(self._datas_hora):
            raise IndexError("Índice inválido")
        return self._datas_hora[indice].isPassado()
    
    def __str__(self) -> str:
        return "\n".join(f"Evento: {self._titulo}, Data: {dh.data_hora}, Descrição: {self._descricao}, Concluído: {dh.isPassado()}" for dh in self._datas_hora)
    
    def editar_data_hora(self, data_hora_antiga: str, data_hora_nova: str):
        for dh in self._datas_hora:
            if dh.data_hora == data_hora_antiga:
                dh.data_hora = data_hora_nova
                return
        raise ValueError(f"Data/hora {data_hora_antiga} não encontrada")

if __name__ == "__main__":
    lista_eventos = [
        EventoUnico("Reunião Importante", "Discussão de projeto", "20/06/2024, 14:00"),
        EventoRecorrente("Reunião Semanal", "Atualização de status", "01/06/2024, 09:00", "31/12/2024, 09:00", 7),
        EventoUnico("Workshop", "Treinamento de novas tecnologias", "15/07/2024, 10:00"),
        EventoRecorrente("Revisão Mensal", "Análise de métricas", "05/06/2024, 16:00", "05/12/2024, 16:00", 30)
    ]

    print("=== AGENDA COMPLETA ===")
    for evento in lista_eventos:
        print(evento)
        print("-" * 60)
````
